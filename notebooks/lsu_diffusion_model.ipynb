# A one-dimentional diffusion model

Here we develop a one-dimentional model of diffusion.
It assumes a constant diffusivity.
It uses a regular grid.
It has a step function for an initial condition.
It has fixed boundary conditions.

The diffusion eqaution:

$$ \frac{\partial C}{\partial t} = D\frac{\partial^2 C}{\partial x^2} $$


The discretized version of the diffusion equation we'll sole with out model:

$$ C^{t+1}_x = C^t_x + {D \Delta t \over \Delta x^2} (C^t_{x+1} - 2C^t_x + C^t_{x-1}) $$

This is the FTCS scheme described in varous textbooks.
(Also Wiki)

We'll use two libraries, Numpy (for  arrays) and Matplotlib (for plots), that are not a part of the core Python library.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Start by setting two fixed model parameter,
the diffusivity and the size of the model domain.

```python
D = 100
Lx = 300
```

Next, set up the model grid with a Numpy Array.

```python
dx = 0.5
x = np.arange (start=0, stop=Lx, step = dx)
nx = len(x)
```

Set up the initial conditions for the model.
The concentration 'C' is a step function with a high value on the left,
a low value on the right, 
and teh step at the center of the domain.

```python
C = np.zeros_like(x)
C_left = 500
C_right = 0
C[x <= Lx//2] = C_left
C[x > Lx//2] = C_right
```

Plot the initial concentration profile.

```python
plt.figure()
plt.plot(x,C,"r")
plt.xlabel("x")
plt.ylabel("C")
plt.title("Initial Concentration Profile")
```

Set up number of time steps in the model.
Calculate a astable times step using a stability criterion.

```python
nt = 5000
dt = 0.5 * dx**2 /D
```

```python
dt
```

Loop over the time steps of the model,
solving the diffusion equation usign the FTCS scheme described above. 
Note the use of array operatons on the concentration 'C'.
The boundary conditions remain fixed in each time step.

```python
for t in range(0, nt):
	C[1:-1] += D * dt / dx ** 2 * (C[:-2] - 2*C[1:-1] + C[2:])
```

Plot the results.

```python
plt.figure()
plt.plot(x,C,"b")
plt.xlabel("x")
plt.ylabel("C")
plt.title("Final Concentration Profile")
```
